```python
..............Question 1.................

for loop is used to run a block of code for a certain number of times.
example:- fruits = ["apple", "banana", "cherry"]
for x in fruits:
    print(x)
while loop is used to run a block code until a certain condition is met.
example:- i = 1
while i < 6:
    print(i)
    i += 1
```

```python
..............Question 4...............

# for loops

number = []
cubedivisible = []

for i in range(1,101):
    number.append(i)
    cube = i**3
    if cube %4==0 or cube%5==0:
        cubedivisible.append(cube)
print(cubedivisible)


# while loops

numbers = []
cubedivisible = []
i = 1

while i <= 100:
    numbers.append(i)
    cube = i ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        cubedivisible.append(cube)
    i += 1

print(cubedivisible)
```
